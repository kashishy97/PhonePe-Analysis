### EXTRA TASK - Trying to fill in the missing data.

```python
# Filling in the missing value in Amount (INR) column in state_txn_users dataset.

# Using interpolate to fill missing value in Amount (INR) after grouping by state.

state_txn_users['Amount (INR)'] = state_txn_users.groupby('State')['Amount (INR)'].transform(lambda x: x.interpolate())

# Filling the value in ATV (INR) based on the new value in Amount (INR).

state_txn_users['ATV (INR)'] = state_txn_users['Amount (INR)']/state_txn_users['Transactions']

# Checking to see if everything worked properly.

state_txn_users[state_txn_users['State'] == 'Andhra Pradesh']
```

```python
# Filling the missing values in ATV (INR) column of distrit_txn_users dataset.

# Using fillna function so that whenever an error pops up, 0 gets filled as the value instead of NaN.

district_txn_users['ATV (INR)'] = (district_txn_users['Amount (INR)']/district_txn_users['Transactions']).fillna(0)

# Checking if the fix worked.

district_txn_users.isnull().any()
```

```python
# We have another column with nulls but since it is impossible to fix them, we will fill them with NA instead.

# Filling missing values in Code column of district_txn_users dataset with 'NA'.

district_txn_users['Code'] = district_txn_users['Code'].fillna('NA')

# Checking to make sure there are no more missing values in the dataset.

district_txn_users.isnull().any().any()
```

# TASK 3 - DATA QUALITY CHECKS

## 3.1 : Ensure data consistency across state and district levels

### 3.1.1 : For each state, calculate the total number of transactions, total transaction amount and total registered users by summing up the values from the district level data.

```python
district_txns = district_txn_users.groupby(['State', 'Year', 'Quarter'])[['Transactions', 'Amount (INR)', 'Registered Users']].sum().reset_index()
district_txns
```

### 3.1.2 : Compare the results with the corresponding values at the state level to ensure they match.

```python
# Creating a df which shows only the required columns.

state_txns = state_txn_users[['State', 'Year', 'Quarter', 'Transactions', 'Amount (INR)', 'Registered Users']]

# Creating a merged df to make it easier to match values.

merged = district_txns.merge(state_txns,
                             on = ['State', 'Year', 'Quarter'],
                             suffixes = ('_d', '_s'))

# Creating new columns to show if values are a match or not.

merged['transactions_match'] = merged['Transactions_d'] == merged['Transactions_s']

# Using the isclose function from numpy since float values can have very miniscule differences

merged['amount_match'] = np.isclose(merged['Amount (INR)_d'],
                                    merged['Amount (INR)_s'],
                                    rtol = 1e-5)                  # Relative tolerance = 0.001%

merged['users_match'] = merged['Registered Users_d'] == merged['Registered Users_s']

# Final comparison df

comparing_df = merged[['State', 'Year', 'Quarter', 'transactions_match', 'amount_match', 'users_match']]
comparing_df
```

### 3.1.3 : Display the discrepancies found between the district level and state level data.

```python
discrepancies = merged[(~merged['transactions_match']) | (~merged['amount_match']) | (~merged['users_match'])]

# ~ = bitwise not operator

discrepancies = discrepancies[['State', 'Year', 'Quarter',
                               'Transactions_d', 'Transactions_s',
                               'Amount (INR)_d', 'Amount (INR)_s',
                               'Registered Users_d', 'Registered Users_s']]

mismatch_counts = {
    'transactions_mismatches': (~merged['transactions_match']).sum(),
    'amount_mismatches': (~merged['amount_match']).sum(),
    'registered_users_mismatches': (~merged['users_match']).sum()
}

print(mismatch_counts)

discrepancies

# From the output, we can see there is only 1 discrepancy in the Amount (INR) column and
# none in the Transactions and Registered Users columns.
```

#### According to our analysis, the 'Transactions' and 'Registered Users' data is consistent between the State and District datasets. There is 1 case of inconsistency in the 'Amount (INR)' column between the datasets.

```python
state_txn_users['Quarter - Year'] = (state_txn_users['Year'].astype(str) + ' Q' + 
                                              state_txn_users['Quarter'].astype(str))

transaction_data = state_txn_users.groupby(['Year', 'Quarter', 'Quarter - Year'])[['Transactions', 'Amount (INR)']].sum().reset_index()

# Transaction Volume plot

plt.figure(figsize = (12, 6))
plt.plot(transaction_data['Quarter - Year'], transaction_data['Transactions'], marker = 'o')

plt.title('Transaction Volumes over time.')
plt.xlabel("Time Frame")
plt.ylabel('Transaction Volume')
plt.grid(linestyle = '--', alpha = 0.5)
plt.show()

# Transaction amount plot

plt.figure(figsize = (12, 6))
plt.plot(transaction_data['Quarter - Year'], transaction_data['Amount (INR)'], marker = 'o')

plt.title('Transaction Amounts over time.')
plt.xlabel("Time Frame")
plt.ylabel('Transaction Amount')
plt.grid(linestyle = '--', alpha = 0.5)
plt.show()
```

#### Insights - 

1. Transactions volume grew from 134 million to 3.9 billion
2. Transaction Amount grew from 171 billion to 7.47 trillion
3. There was a significant jump in both volume and amount after Q2 2020, likely due to post-lockdown recovery.
